### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def e_greedy_policy(env, Q, epsilon, state):
    greedy_prob = 1-epsilon
    if np.random.random() < greedy_prob:
        action = np.argmax(Q[state])
    else:
        action = env.action_space.sample()
    return action

def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        ## TODO: complete the function       
#         epsilon = (1-(const_ep_decay*i_episode)) if ((1-(const_ep_decay*i_episode)) > 0.1) else 0.1 # as described in the alpha go paper
#         epsilon = 1/i_episode if 1/i_episode > 0.1 else 0.1
        epsilon = 1/i_episode
        state = (env.reset(),)
        action = e_greedy_policy(env, Q, epsilon, state[0])
        while True:
            next_state, reward, done, prob = env.step(action)
            next_action = e_greedy_policy(env, Q, epsilon, next_state)
            Q[state][action] += alpha*((reward + gamma*Q[next_state][next_action])-Q[state][action])
            action = next_action
            state = next_state
            if done:
                break
        
    return Q
````

### Part 3: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def e_greedy_policy(env, Q, epsilon, state):
    greedy_prob = 1-epsilon
    if np.random.random() < greedy_prob:
        action = np.argmax(Q[state])
    else:
        action = env.action_space.sample()
    return action

def expected_state_value(Q, state, epsilon, env):
    max_prob = (1-epsilon) + epsilon/env.nA
    non_max_prob = epsilon/env.nA
    max_action_value = np.max(Q[state])*max_prob
    argmax = np.argmax(Q[state])
    non_max_action_values = 0
    for i in range(env.nA):
        if i != argmax:
            non_max_action_values += non_max_prob*Q[state][i]
    return max_action_value + non_max_action_values
    
def expected_sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        ## TODO: complete the function       
        epsilon = 1/i_episode
        state = (env.reset(),)
        action = e_greedy_policy(env, Q, epsilon, state[0])
        while True:
            next_state, reward, done, prob = env.step(action)
            Q[state][action] += alpha*((reward + gamma*expected_state_value(Q, next_state, epsilon, env))-Q[state][action])
            action = e_greedy_policy(env, Q, epsilon, next_state)
            state = next_state
            if done:
                break
        
    return Q
````
